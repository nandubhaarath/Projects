```python
minvalue = lengths.min()
possibleBoundaries = np.arange(lengths.min() , lengths.max()+0.15 , step=0.1)
for boundary in possibleBoundaries:

  virginica_pred =  (lengths >= boundary).nonzero()[0]
  virginca_true = (virginica_pred >= 50).nonzero()[0]

  versicolour_pred =  (lengths < boundary).nonzero()[0]
  versicolour_true = (versicolour_pred <= 49).nonzero()[0]

  accuracy = (len(virginca_true) + len(versicolour_true))
  print("At" , boundary.round(1) , "accuracy is" , accuracy)
  #calculating the accuracy values
```
